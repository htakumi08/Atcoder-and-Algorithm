```python
# B-o 7分くらい
# 比較的スムーズにいけた。演習の成果が少しはでている気がする
N,Q = map(int,input().split())
T = list(map(int,input().split()))

n_cnt = [0]*N
for i in range(N):
    n_cnt[i] += 1 

for i in range(Q):
    if n_cnt[T[i]-1] == 1:
        n_cnt[T[i]-1] -= 1
    else:
        n_cnt[T[i]-1] += 1

print(sum(n_cnt))
```

C問題  https://www.youtube.com/watch?v=47DxnXrmjg4  
愚直な全探索でのソート→O(N**2)で計算量over  
→1回の探索1→Nで、配列の要素を1→Nに整列させる。  
→要素を探索で探すのではなく、住所録を作ってそこを参照して整列していく  
→A[3,2,4,1] : pos[4,2,1,3] (1→Nが配列Aのどのindexにあるのか保存)  
→posを確認してAを書き換え、その後のposも書き換える.

```python
# C-x
# これは絶対に解けていい問題だった。
"""
自分が思いついた一つの解法にすがりつきすぎ。
→ 違ったならば誤ってる事実を受け入れ何が違うのかどこが違うのかを精査しつづけろ
→ なんとなく類題っぽいのを解いた記憶があるから、一度同じような解法を見かけたならば次は絶対に解けるようにしろ
→ 解法じゃなくて、どういう思考プロセスをたどれば解法にたどり着くのかを理解しろ
"""
N = int(input())
A = list(map(int,input().split()))

"""思考順
1. 入れ替え数cnt、A[i]とA[A[i]]の入れ替え ex.i=1,A[i]=3 → j=A[i]=3とを入れ替える
    (1→これだと正しく配列Aを1~Nで昇順にソートできない場合がある)
2. WAだった→「ソートしてない場合のこと考えないといけないのか？」
→ 一度でも入れ替えたらresult=Trueとかやってみる.

"""
cnt = 0
ans = []
result = False
for i in range(N):
    if A[i] != i+1:
        result = True
        cnt += 1
        tmp = A[i] - 1
        A[i],A[tmp] = A[tmp],A[i]
        ans.append([i+1,tmp+1])

print(cnt)
for i in range(len(ans)):
    if result == False:
        print(0)
        break
    
    print(*sorted(ans[i]))
```

```python
# C-解答
N = int(input())
# 1-indexで考えれるようにしとくと、操作や処理で間違いづらい
A = [0] + list(map(int,input().split()))
pos = [0]*(N+1)
for i in range(1,N+1):
    pos[A[i]] = i

ans = []
for i in range(1,N+1):
    # 配列Aが1~Nでと整理されてないならば
    if A[i] != i:
        tmp = pos[i]
        # i,tmpの位置を入れ替える→入れ替え位置をansに保存
        ans.append(min(i,tmp),max(i,tmp))   
        A[i],A[tmp] = A[tmp],A[i]
        # positionも入れ替える
        pos[A[tmp]] = tmp
        pos[A[i]] = i

print(len(ans))
for a in ans:
    print(*a)
```
