```python
# リストにあるすべての0を末尾に移動

def move_zeros(a_list):
    zero_index = 0
    for index, num in enumerate(a_list):
        if num != 0:
            a_list[zero_index] = num
            if zero_index != index:
                a_list[index] = 0
            zero_index += 1
    return a_list

a_list = [8,0,3,0,12]
print(a_list)
move_zeros(a_list)
print(a_list)
```

```python
# リスト内の重複を見つける
# dup重複
def return_dups(an_iterable):
    dups = []
    a_set = set()
    for item in an_iterable:
        l1 = len(a_set)
        a_set.add(item)
        l2 = len(a_set)
        if l1 == l2:
            dups.append(item)
    return dups

a_list = [
    "Susan Adams",
    "Kwame Goodall",
    "Jill Hampton",
    "Susan Adams",      
]
print(a_list)
print(set(a_list))
print("--------\n")

dups = return_dups(a_list)
print(dups)
```

```python
# 2つのリストの交差を探す
    # in リスト内要素を線形探索　O(n)
list1 = [2,43,48,62,64,28,3]
list2 = [1,28,42,70,2,10,62,31,4,14]

def return_inter(list1, list2):
    list3 = [v for v in list1 if v in list2]
    return list3

print(return_inter(list1, list2))

```
