```python
# C - Pyramid  ABC112
# まったく分からん。まじ分からんな。

# 解答
N = int(input())
x, y, h = [], [], []
for _ in range(N):
    xi, yi, hi = map(int, input().split())
    x.append(xi)
    y.append(yi)
    h.append(hi)

# 高さが非ゼロの点を見つけ、中心を決定するための参考点として使用
tmp = None
for i,hi in enumerate(h):
    if hi >= 1:
        tmp = i
G = (x[tmp],y[tmp],h[tmp])

ans = []

# 中心となりうるすべての座標を探索
for i in range(101):
    for j in range(101):
        # 参考点からの距離を考慮した高さの計算
        H = G[2] + abs(G[0]-i) + abs(G[1]-j)
        H = max(H,0)

        # 各点について計算された高さが一致するかを確認
        # i,jが中心座標だとしたら
        valid = 0
        for k in range(N):
            H_cur = H - abs(x[k]-i) - abs(y[k]-j)
            H_cur = max(H_cur,0)
            if h[k] != H_cur:
                valid = False
                break
        
        # すべての点の高さが一致しているなら答えリストに追加
        if valid:
            ans.appned((i,j,H))

# 唯一の中心が見つかったときにそれを出力
if len(ans) == 1:
    print(ans[0][0],ans[0][1],ans[0][2])
```

# C問題　Greedyアルゴリズム,端から決まって解くGreedy,後ろから解く
    

```python
# C - 白昼夢  ABC49

# 解答
S = input()
divide = ["dream","dreamer","erase","eraser"]

# 入力文字列とdivideの各文字列を反転
S = S[::-1]
for i in range(4):
    divide[i] = divide[i][::-1]

# 文字列を後ろから調べる
can = True
i = 0
for _ in range(len(S)):
    # ４つの文字列のどれかでdivideできるかの確認フラグ
    can2 = False
    for d in divide:
        # 現在の位置から部分文字列がdivideのいずれかと一致するか確認
        if S[i:i+len(d)] == d:
            can2 = True
            # dividできたのでi進める
            i += len(d)
            break
    
    # どの文字列にも一致しなかった場合
    if not can2:
        can = False
        break
    # 文字列の最後まで到達した場合
    if i == len(S):
        break

print('YES' if can else 'NO')
```

```python
# A - Sorted Arrays  AtCoder Grand Contest 013
# いや、わからんわ。
N = int(input())
A = list(map(int, input().split()))

# 初期状態をセット（0:初期、1:増加、-1:減少）
prev_dir = 0
count = 0

i = 0
# 初期状態の場合方向を決める
while i < N:
    if prev_dir == 0:
        if i < N-1 and A[i] == A[i+1]:
            i += 1
            continue
        elif A[i] < A[i+1]:
            prev_dir = 1
        elif A[i] > A[i+1]:
            prev_dir = -1
        # N-2とN-1とで区切られて、残りN-1のみを考える場合
        elif i == N-1:
            i += 1
            count += i
    
    # 増加方向の場合
    if prev_dir == 1:
        while i < N-1 and A[i] <= A[i+1]:
            i += 1
        count += 1
        prev_dir = 0
    # 減少方向の場合
    elif prev_dir == -1:
        while i < N-1 and A[i] >= A[i+1]:
            i += 1
        count += 1
        prev_dir = 0
    
    i += 1
    
print(count)
```

```python
# C - Sequence  59
n = int(input())
a = list(map(int, input().split()))
cs = []

# 累積和の計算
# 返り値ndarray np.cumsum(array-likeオブジェクト)
for i in range(n):
    if i == 0:
        cs.append(a[0])
    else:
        cs.append(a[i] + cs[i-1])

#️ 奇数番目正
# num:累積和の修正量　→　次の累積和にも影響を与える。
ans1 = 0
num = 0
for i in range(n):
    # i：偶数のとき
    if i % 2 == 0 and cs[i] + num <= 0:
        ans1 += -(cs[i]+num) + 1
        num += -(cs[i]+num) + 1
    # i:奇数のとき
    if i % 2 and cs[i] + num >= 0:
        ans1 += cs[i] + num + 1
        num -= cs[i] + num + 1

# 奇数番目負
ans2 = 0
num = 0
for i in range(n):
    #️ i：偶数のとき
    if i % 2 and cs[i] + num <= 0:
        ans2 += -(cs[i] + num) + 1
        num += -(cs[i] + num) + 1
    # i:奇数のとき
    if i % 2 == 0 and cs[i] + num >= 0:
        ans2 += cs[i] + num + 1
        num -= cs[i] + num + 1

print(min(ans1,ans2))
```
