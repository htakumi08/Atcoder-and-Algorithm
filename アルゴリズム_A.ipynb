```python
def binary_search(a_list, key):
    #配列a_listの左端、右端
    left,right = 0, len(a_list)-1

    # 探索範囲がある限り
    while right >= left:
        # 整数値大きくなりオーバーフロー防ぐ工夫
        # 探索範囲の幅の半分を求め、現在の探索範囲の左端lefからその中央までの距離を加えるという意味
        mid = left + (right-left)//2
        if a_list[mid] == key:
            return mid
        elif a_list[mid] > key:
            right = mid - 1
        else:
            left = mid +1
    return -1

a = [1,14,32,51,51,243,419,750,910]

print(binary_search(a, 51))
print(binary_search(a, 52))
print(binary_search(a, 243))
```

```python
# より一般化
def isOK(index,a,key):
    return a[index] >= key

# ok:条件満たす可能性のある最小のindex ex.配列外の大きさの数値a[2,3,78,] key=834, ok=3 
def binary_search(a,key):
    ng = -1
    ok = len(a)

    # abs(ok-ng)探索の範囲が動的に変化する場合:ex.前から探索か、後ろから探索か
    # 前後の探索関係なくこの関数で求めれる一般化
    while abs(ok-ng) > 1:
        mid = (ok+ng)//2
        if isOK(mid, a, key):
            ok = mid
        else:
            ng = mid
    return ok
```

```python
# A12
# 答えに対して二分探索
N,K = map(int,input().split())
A = list(map(int,input().split()))

# A秒あたりにどのくらい各プリンターがチラシを印刷できるか
# そのチラシの合計と求めるK枚に対する大小比較
def check(mid,N,K,A):
    sum = 0
    for i in range(N):
        sum += mid // A[i]
    if sum >= K:
        return True
    return False

# 二分探索
# A秒あたり→最大値の10*9を考える
left = 1
right = 10**9
while left < right:
    mid = (left+right)//2
    ans = check(mid,N,K,A)
    if ans == False:
        # ooo(mid)ooo...でmid以下で条件Falseなのでmid+1をleftにする-★
    	left = mid + 1
    else:
        ans == True
    	right = mid
# 条件left<rightと★から求めたい秒数はleftに保存される
print(left)
```

```python
# 別解
def check(x,N,K,A):
    sum = 0
    # x秒までに各Aiが何枚印刷するかの合計を求める
    for i in range(N):
        sum += x//A[i]

    if  sum >= K:
        return True
    return False

ng = -1
ok = 10**9
while abs(ok-ng) > 1:
    mid = (ok+ng)//2
    ans = check(mid,N,K,A)
    if ans:
        ok = mid
    else:
        ng = mid
print(ok)
```

```python
# A13
# しゃくとり法：https://qiita.com/drken/items/ecd1a472d3a0e7db8dce
# 条件を満たす区間をすべて求める事ができる→区間のうちmin,maxの長さや数え上げ
N,K = map(int,input().split())
A = list(map(int,input().split()))

idx_count = [0]*N

# しゃくとり法→配列などの連続する要素対して条件満たす部分列を効率的に探索するアルゴリズム
# ソートされている配列などに有効か
# 配列Aの要素数より１つ小さく考えないと比較できなくなるよ
for i in range(N-1):
    if i == 0:
        idx_count[i] = 0
    else:
        # 開始点が一つ右に移動したときにそれに伴い終点も同じかより右になることが保証されている.
        idx_count[i] = idx_count[i-1]
	# idx_count[i]には下記の条件満たす配列Aの右側（終端）のインデックスを格納する.
    while idx_count[i] < N-1 and A[idx_count[i]+1]-A[i] <= K:
        idx_count[i] += 1

ans = 0
# 配列Aのi番目の要素が求める条件満たすのは、配列Aのidx_count[i]番目の要素まで.
for i in range(N-1):
    ans += (idx_count[i]-i)
print(ans)
```

```python
# 簡単なDPの例
N = int(input())
A = list(map(int,input().split()))
# 入力N=3,A=[7,-6,9]

# DPテーブル
dp = [0]*(N+1)

# 動的計画法→ノードの値を順々に更新していく手法
# ノード:データ構造やネットワークにおいてデータやオブジェクトを表す基本的な要素のこと
# dp[0]:何も選ばない,dp[1]:A[0]の中から選ぶ,dp[i+1]:A[0:i]から選ぶ
for i in range(N):
    # dp[i]までの最大値→A[i]を選ぶか選ばないか→二択のうち最大値取る方を選ぶ
    dp[i+1] = max(dp[i], dp[i]+A[i])

print(dp[N])
```

```python
# A17
# 最短経路を求める.
N = int(input())
A = list(map(int,input().split()))
B = list(map(int,input().split()))

# 動的計画法
dp = [0]*(N+1)
dp[2] = A[0]
for i in range(3,N+1):
    dp[i] = min(dp[i-1]+A[i-2], dp[i-2]+B[i-3])

# 答えの復元
# 変数placeは現在位置(ゴールから逆算してく)
ans = []
place = N
while True:
    ans.append(place)
    if place == 1:
        break
    if dp[place] == dp[place-1]+A[place-2]:
        place -= 1
    else:
        place -= 2
ans.reverse()

print(len(ans))
print(*ans)
```

```python
# A18
# DPの部分和問題
# 与えられた整数の集合からいくつかの数を選び和が特定のSになるかどうかを判断する
N,S = map(int,input().split())
A = list(map(int,input().split()))

# 動的計画法(i=0)
# dp[N][S]テーブル:N個の整数で和Sが作れるかどうかの真偽値を格納
# 最初のi個の整数(A[0]~A[i-1])を使って和jを作ることができるかどうか真偽値を格納
dp = [[False]*(S+1) for _ in range(N+1)]
# 0個の整数を使って和0を作る→可能True
dp[0][0] = True

# 動的計画法(i=1)
for i in range(1,N+1):
	# 0から始めることで、j < A[i-1] の場合もカバ
	for j in range(S+1):
		if j <= A[i-1]:
			if dp[i-1][j] == True:
				dp[i][j] = True
		if j > A[i-1]:
			if dp[i-1][j-A[i-1]] == True:
				dp[i][j] = True

print('Yes' if dp[N][S] else 'No')
```

```python
#A53
import heapq

# 入力
Q = int(input()) # クエリ数
queries = [ input().split() for i in range(Q) ]

# クエリの処理-heapq-優先度付きキュー
T = []
for q in queries:
    if q[0] == "1":
        heapq.heappush(T,int(q[1]))
    elif q[0] == "2":
        # T[0]は優先度付きキューの最小要素になっている.
        print(T[0])
    elif q[0] == "3":
        # 最小要素を取り出す ⇔ heap[0]
        heapq.heappop(T)

"""
注意1: Python の heapq モジュールは、deque モジュールとは異なり
関数を使って list を操作するという形式になっています。
注意2: 優先度付きキューの最小要素は T[0] で取り出せますが、
例えば小さいほうから 2 番目の要素が T[1] で取り出せるとは限らないことに注意してください
"""
```

```python
# A56
# 連続した文字列の部分文字列のハッシュ値の効率的な計算
# ローリングハッシュO(N+M): 文字列検索を高速で行うアルゴリズム
"""
https://qiita.com/hirominn/items/80464ee381c8d400725f
https://qiita.com/keymoon/items/11fac5627672a6d6a9f6
"""

# 愚直に探索すると計算量オーバー

N, Q = map(int, input().split())
S = input()
queries = [ list(map(int, input().split())) for i in range(Q) ]

# 文字を数値に変換
# ord(x)である文字xの文字コード(ASCIIコード)を取得
# lambda:無名関数 lambda 引数:評価する式
# "a"からの相対位置をTに格納
T = list(map(lambda x: ord(x)-ord("a") +1,S))
"""
abcbabc
T:[1, 2, 3, 2, 1, 2, 3]
"""

# 100 の n 乗を前計算
# MOD = 2**31 -1 32ビット整数の最大値であり素数
# 十分大きいMODの使用でハッシュ衝突(余りが同じ値になる)ことを防げる.
# MOD(素数)で割ることでハッシュ値の最大桁数を決められる。
MOD = 2147483647
power100 = [ None ] * (N + 1)
power100[0] = 1
for i in range(N):
	power100[i + 1] = power100[i] * 100 % MOD
"""power100:[1, 100, 10000, 1000000, 100000000, 1410065412, 1420104145, 276493798]"""

# ハッシュ値を作る → 元データから算出される値
# 文字列Sの先頭から各位置までの部分文字列のハッシュ値を計算→Hに格納
H = [ None ] * (N + 1)
H[0] = 0
for i in range(N):
	H[i + 1] = (H[i] * 100 + T[i]) % MOD
"""H:[0, 1, 102, 10203, 1020302, 102030201, 1613085514, 247277878]"""

# ハッシュ値を求める関数
# S[l-1:r] のハッシュ値は (H[r] - H[l - 1] * power100[r - l + 1]) % MOD で計算
# C++ とは異なり、（負の値）% M (M >= 1) も 0 以上 M-1 以下になることに注意
def hash_value(l, r):
	return (H[r] - H[l - 1] * power100[r - l + 1]) % MOD

# クエリに答える
for a, b, c, d in queries:
	hash1 = hash_value(a, b)
	hash2 = hash_value(c, d)
	if hash1 == hash2:
		print("Yes")
	else:
		print("No")
```

```python
# A57
"""
ダブリングO(NlogK)
https://zenn.dev/fjnkt98/articles/3c0c21778b6101
木やリスト等のデータ構造における各要素から、ある特定の条件に基づいて2のべき乗回先の
要素を効率的に探索するアルゴリズム
「最小共通祖先」問題やある状態から特定回数の操作を経て到達する状態を求める問題等に有効
"""

N, Q = map(int, input().split())
A = list(map(int, input().split()))
queries = [ list(map(int, input().split())) for i in range(Q) ]

# 前計算
# 任意のY日後に対して効率的に到達点を求めるための前処理
LEVELS = 30		# Yi日後 <= 10**9より2**level >= Yi_maxとなるような値を設定
dp = [[None]*N for _  in range(LEVELS)]
# d=0→1回のジャンプで到達する要素 インデックス0から数え上げるのでA[i]-1
for i in range(0,N):
    dp[0][i] = A[i]-1
    
# i番目の要素から2^d回ジャンプしたあとに到達する要素を格納
for d in range(1,LEVELS):
    for i in range(0,N):
        dp[d][i] = dp[d-1][dp[d-1][i]]

# クエリ処理
for X,Y in queries:
	# 0indexで考えるのでX-1
    current_place = X-1
    for d in range(29,-1,-1):
        # Yを2進数表記した時の各ビットを調べることで、必要なジャンプ回数を特定
        if ((Y>>d)&1) == 1:
            current_place = dp[d][current_place]
    print(current_place + 1)
```

```python
# 60
"""
計算量:for(N)-while(len(level2))だが、A = N-1,N-2,,,3,2,Nでも
スタックへの追加削除はN-1回 → for1回ループでO(1)処理程度なので全体の計算量O(N)
"""
from collections import deque

N = int(input())
A = list(map(int, input().split()))

# スタックの変化を再現
# スタックには(日付、株価)のタプルを記録する
# deque() → 先頭末尾の処理O(1)で処理可能
"""
先頭以降、先頭保存しi番目とi-1番目を比較
→ i-1番目がi番目より小さい
→ 残りのスタックを都度i番目と比較
→ i<jでiよりもjが大き場合、j以降はi~j-1まで無視していいこと
"""
ans = [None]*N
level2 = deque()
for i in range(N):
    if i >= 1:
        level2.append((i,A[i-1]))
        while len(level2) >= 1:
            # lebel2の末尾から取り出し比較していく.
            # スタック処理
            kabuka = level2[-1][1]  # 株価はタプルの2番目の要素
            if kabuka <= A[i]:
                level2.pop()
            else:
                break
    if len(level2) >= 1:
        ans[i] = level2[-1][0]
    else:
        ans[i] =-1

print(*ans)
```

```python
# A62-dfs
# pythonの再帰呼び出し深さはdefaultで1000まで 
# 十分大きい値を設定→無理なら少し増やしていく
import sys
sys.setrecursionlimit(120000)

N,M = map(int,input().split())
edges = [list(map(int,input().split())) for _ in range(M)]

G = [[] for _ in range(N+1)] 
for a,b in edges:
    G[a].append(b)
    G[b].append(a)

def dfs(G,v,visited):
    visited[v] = True
    for next_v in G[v]:
        if visited[next_v]:
            continue

        dfs(G,next_v,visited)

visited = [False]*(N+1)
dfs(G,1,visited)

result = True
for i in range(1,N+1):
    if visited[i] == False:
        result = False
        break
# result = all(visited[1:]) # 1番目の頂点からN番目の頂点すべての要素がTrueかどうか

print("The graph is connected." if result else "The graph is not connected.")
```

```python
# A63-bfs
# 始点から各頂点への最短経路を求めることが得意
from collections import deque

N,M = map(int,input().split())
edges = [list(map(int,input().split())) for i in range(M)]

G = [[] for _ in range(N+1)]
for a,b in edges:
    G[a].append(b)
    G[b].append(a)

dist = [-1]*(N+1)
dist[1] = 0
que = deque()
que.append(1)

while len(que) >= 1:
    # キューの先頭要素を取り除きvに代入
    v = que.popleft()
    for next_v in G[v]:
        if dist[next_v] == -1:
            dist[next_v] = dist[v] + 1
            que.append(next_v)

# 頂点１から各頂点までの最短距離を出力
for i in range(1,N+1):
    print(dist[i])
```

```python
# 64-ダイクストラ(bfsっぽい

# 重み付き(優先度付き)の最短経路問題
# heapq:重み付きのqueue
import heapq

N, M = map(int, input().split())
edges = [ list(map(int, input().split())) for i in range(M) ]

# 隣接リストの作成（重み付きグラフなので、各辺について (隣接頂点, 重み) のタプルを記録する）
G = [ list() for i in range(N + 1) ]
for a, b, c in edges:
    G[a].append((b,c))
    G[b].append((a,c))

# 配列・キューの初期化：キューには（距離、頂点番号）のタプルを記録
INF = 10**10
kakutei = [False]*(N+1)
# cur[x]:頂点xに対する最短経路を格納
cur = [INF]*(N+1)
cur[1] = 0
Q = []
heapq.heappush(Q,(cur[1],1))

# ダイクストラ法
"""ある頂点posに対して最短経路が確定したあと,posから隣接するすべての頂点の最短経路を更新し、
   その情報を優先度キューQに追加する"""
while len(Q) >= 1:
    # 次に確定させるべき頂点を求める
    # ここでは、優先度付きキューQの最小要素を取り除き、その要素の２番目の値（頂点番号）をposに代入する
    # Qのタプル(距離(優先度)、頂点番号)の距離が最小のものをpopしそのタプルの頂点番号をposに代入
    pos = heapq.heappop(Q)[1]

    # Qの最小要素が「既に確定した頂点」の場合
    if kakutei[pos] == True:
        continue
    
    # cur[x]の値を更新する
    # 頂点posまでの最短経路の確定
    kakutei[pos] = True
    for e in G[pos]:
        # e:タプル→e[0]頂点,e[1]posからe[0]までの重み（距離）
        # cur[頂点]にはこの頂点までの現在の最短経路用に最小値を格納したい→あらかじめ十分大きい値で初期化
        if cur[e[0]] > cur[pos] + e[1]:
            cur[e[0]] = cur[pos] + e[1]
            heapq.heappush(Q,(cur[e[0]],e[0]))


for i in range(1,N+1):
    if cur[i] != INF:
        print(cur[i])
    else:
        print("-1")
```

```python
# A65 - 木構造
# https://www.cqpub.co.jp/hanbai/books/18/18781/18781_9syo.pdf

# A[2]....A[N]となるようにする
N = int(input())
# 社長(root)をA[1]で考えられるように.
A = [0]*2 + list(map(int,input().split()))
print(A)

# 隣接リスト
# 上司Ai→部下iの方向に辺を追加
G = [[] for i in range(N+1)]
for i in range(2, N+1):
    G[A[i]].append(i)
print(G)

# 動的計画法:dp[x]は社員xの部下の数
dp = [0]*(N+1)
for i in range(N,0,-1):
    # 木構造の葉の部分(木の終端)はスルー
    for j in G[i]:
        # dp[j]+1 →抱えてる部下の数＋自分自身
        dp[i] += (dp[j] + 1)

print(*dp[1:])
```

```python
# 66
"""
Union-find:グループ分けを効率的に管理できるデータ構造:根付き木の構造
統合クエリ:要素uを含むグループと要素vを含むグループを統合する処理
回答クエリ:要素uと要素vが同じグループにあるかどうかを答える処理
"""

N,Q = map(int,input().split())
queries = [list(map(int,input().split())) for i in range(Q)]

# Union-Find木クラス
class UnionFind:
    # n頂点のunion-find木を作成
    def __init__(self,n):
        self.n = n
        # 1indexで直感的に扱えるようにする(頂点番号1~Nのため)-★
        self.par = [-1]*(n+1)   # 最初は親が無い
        self.size = [1]*(n+1)   # 最初はグループの頂点数が1
        # self.size:頂点iを根とする木の頂点数

    # 頂点xの根を返す関数
    # ★1indexによりxで扱う. 0indexの場合はx-1等で調節する
    def root(self,x):
        while self.par[x] != -1:
            x = self.par[x]
        return x

    # 統合クエリ:要素u,vを統合する関数
    def unite(self,u,v):
        rootu = self.root(u)
        rootv = self.root(v)
        if rootu != rootv:     # この処理無いとTLE → 全頂点について調べることになるから 
            # uとvが異なるグループのときのみ処理を行う
            if self.size[rootu] < self.size[rootv]:
                self.par[rootu] = rootv
                self.size[rootv] += self.size[rootu]
            else:
                self.par[rootv] = rootu
                self.size[rootu] += self.size[rootv]
    
    # 回答クエリ:要素uとvが同一のグループかどうかを返す関数
    def same(self,u,v):
        return self.root(u) == self.root(v)

# クエリの処理
uf = UnionFind(N)
for tp,u,v, in queries:
    if tp == 1:
        uf.unite(u,v)
    if tp == 2:
        if uf.same(u,v):
            print("Yes")
        else:
            print("No")
```

### DFS(深さ優先探索)-一般型  
https://qiita.com/drken/items/4a7869c5e304883f539b  
計算量(頂点数＋辺数):頂点Nを一通り調べるO(N)　辺数を一通り調べるO(M)　２つの線形時間  
スタックの処理

```python
# 一般型
def dfs(G,v,seen):
    seen[v] = True  # 頂点vを訪問済にする

    # vから行ける拡張点next_vについて調べる
    for next_v in G[v]:
        if seen[next_v]:    # next_vが探索済だったらスルー
            continue

        dfs(G,next_v,seen)  # 再帰的に探索
    
N,M = map(int,input().split())

# グラフの入力受取(無向グラフ想定)
G = [[] for _ in range(N)]  # 隣接リス表現
for _ in range(M):
    a,b = map(int,input().split())
    G[a].append(b)
    G[b].append(a)

# 頂点0をスタートとした探索
seen = [False] * N  # 全頂点を「未訪問」に初期化
dfs(G,0,seen)
```

### BFS(幅優先探索)-一般型  
https://qiita.com/drken/items/996d80bcae64649a6580  
計算量(頂点数＋辺数):線形時間  
最短経路アルゴリズムに向く  
キューの処理

```python
# 一般型
from collections import deque

# 頂点数と辺数の入力受取
N, M = map(int, input().split())

# グラフの入力受取(無効グラフを想定)
G = [[] for _ in range(N)]
for _ in range(M):
    a,b = map(int,input().split())
    # 頂点aと頂点bを繋いでいる
    G[a].append(b)
    G[b].append(a)

# BFSのためのデータ構造
# 全頂点を「未訪問」に初期化
dist = [-1]*N
que = deque()

# 初期条件(頂点0を初期ノードとする)
dist[0] = 0
que.append(0)   # 頂点0 を発見済とする

# BFS開始(キューが空になるまで探索を行う)
while que:
    v = que.popleft  # キューから先頭の発見済の頂点を取り出す
    for nv in G[v]:  # 隣接する頂点を探索
        if dist[nv] != -1:  # すでに発見済ならば頂点は探索しない 
            continue

        dist[nv] = dist[v] + 1    # 新たな頂点nvに対して距離情報を更新してキューに追加
        que.append(nv)  # 発見済の頂点を発見済とする

for v in range(N):
    print(f"{v}: {dist[v]}")
```
